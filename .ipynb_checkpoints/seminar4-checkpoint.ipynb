```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('seminar4/4laptop_price.csv', encoding = 'latin-1')
```

```python
df.head()
```

```python
def convert(x):
    res = 0
    memory = str(x).split(' ')[0]
    if memory.endswith('GB'):
        res = memory.replace('GB', ' ')
    elif memory.endswith('TB'):
        res = memory.replace('TB', ' ')*1024
    return res
df['Memory_Amount'] = df['Memory'].apply(convert)

```

```python
plt.hist(df['Memory_Amount'])
plt.title('Распределение памяти')
plt.xlabel('GB')
plt.ylabel('Количество')
plt.xticks(rotation=30)
```

```python
plt.hist(df['Price_euros'])
plt.title('Распределение стоимости')
plt.xlabel('euro')
plt.ylabel('Количество')
```

```python
df['Weight'] = df['Weight'].str.replace('kg', '')
df['Weight'] = df['Weight'].astype('float')
plt.hist(df['Weight'])
plt.title('Распределение веса')
plt.xlabel('кг')
plt.ylabel('Количество')
plt.xticks(rotation=30)
```

```python
data = pd.crosstab(index=df['Company'], columns=df['Cpu_Company'])
data = data.drop(columns=['Samsung']).reset_index()
data
```

```python
import numpy as np
plt.figure(figsize=(8,6))
n_ticks=np.arange(len(data['AMD']))
offset=0.2

plt.bar(n_ticks - offset, data['AMD'], width = 0.4)
plt.bar(n_ticks + offset, data['Intel'], width = 0.4)
plt.title('Ноутбуки')
plt.xlabel('Компания')
plt.ylabel('Количество')
plt.legend(['AMD', 'Intel'])
plt.xticks(n_ticks, data['Company'], rotation=30)
```

```python
data = pd.crosstab(index=df['Company'], columns=df['Cpu_Company'], normalize='index')
data = data.drop(columns=['Samsung']).reset_index()
data
```

```python
import numpy as np
plt.figure(figsize=(8,6))

plt.bar(data['Company'], data['AMD'])
plt.bar(data['Company'], data['Intel'], bottom=data['AMD'])
plt.title('Ноутбуки')
plt.xlabel('Компания')
plt.ylabel('Доля ноутбуков')
plt.legend(['AMD', 'Intel'])
plt.xticks(rotation=50)
```

Постройте матрицу корреляций для таблицы

```python
corr_matrix = df.corr()
corr_matrix=np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot = True, linewidth  = .5, cmap = 'coolwarm')
```
