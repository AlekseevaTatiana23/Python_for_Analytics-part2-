## Условие:
## 1. Постройте гистограмму, отображающую распределение цен на недвижимость в наборе данных из файла kc-house-data.
## 2. Постройте гистограмму для анализа распределения площади жилых помещений. На графике должно быть ясно видно, какие значения площади наиболее распространены.
## 3. Постройте гистограмму, показывающую распределение годов постройки домов. График должен помочь выявить тенденции в строительстве по годам.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('kc_house_data.csv', encoding = 'latin-1')
df.head()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['price'], bins=20)
plt.title('Распределение цен на недвижимость')
plt.xlabel('Цена')
plt.ylabel('Количество домов')
plt.grid(True)
```

## Вывод: стоимость большинства домов распределена в диапазоне до 1000000 y.e. Преобладают недорогие дома до 500000 у.е.

```python
plt.figure(figsize=(10, 6))
plt.hist(df['sqft_living'], bins=30)
plt.title('Распределение жилой площади')
plt.xlabel('Жилая площадь (sqft)')
plt.ylabel('Количество домов')
plt.grid(True)
plt.show()
```

## Вывод: жилая площадь большинства домов распределена в диапазоне 1000-2000 фт.кв. Данные площади являются наиболее типичными для выборки домов.

```python
plt.figure(figsize=(10, 6))
plt.hist(df['yr_built'], bins=30)
plt.title('Распределение годов постройки домов')
plt.xlabel('Год постройки')
plt.ylabel('Количество домов')
plt.grid(True)
plt.show()
```

## Вывод: преобладают дома новой постройки. В настоящее время строится больше домов, а также рост строительства наблюдается в 1950-1960 годах. Самый пик строительства приходится на 2000-ые года

# Задача 2. Анализ характеристик домов
## 1. Постройте столбчатую диаграмму для анализа количества домов с видом на набережную и без вида. График должен показать соотношение между домами с видом и без него.
## 2. Постройте столбчатую диаграмму, отображающую распределение домов по количеству этажей. График должен наглядно показать, как часто встречаются дома с разным количеством этажей.
## 3. Постройте столбчатую диаграмму для анализа распределения состояния домов. График должен показывать частоту встречаемости различных состояний домов.


```python
plt.figure(figsize=(8, 6))
data = df['waterfront'].value_counts()
plt.bar(data.index, data.values, color = ['green', 'yellow'])
plt.title('Распределение домов по наличию вида на набережную')
plt.xlabel('Вид на набережную')
plt.ylabel('Количество домов')
plt.xticks([0, 1], ['Нет', 'Да'], rotation=0)
plt.grid(axis='y')
plt.show()
```

## Вывод: очень небольшое количество домов имеет вид на набережную

```python
df['floors'].max(), df['floors'].min()
```

```python
plt.figure(figsize=(8, 6))
data = df['floors'].value_counts().sort_index()
plt.bar(data.index, data.values, color = ['green', 'yellow'])
plt.title('Распределение домов по количеству этажей')
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов')
plt.grid(axis='y')
plt.show()
```

## Вывод: преобладают одноэтажные дома

```python
plt.figure(figsize=(8, 6))
data = df['condition'].value_counts().sort_index()
plt.bar(data.index, data.values, color = ['green', 'yellow'])
plt.title('Распределение домов по состоянию')
plt.xlabel('Состояние')
plt.ylabel('Количество домов')
plt.grid(axis='y')
plt.show()
```

## Вывод: большая часть домов находится в нормальном состоянии, мало домов в плохом состоянии, менее 6000 домов в состоянии выше среднего, менее 2000 домов в отличном состоянии.

## 1. Постройте диаграмму рассеяния, чтобы изучить зависимость между количеством спален и стоимостью недвижимости. График должен показать, как изменение количества спален влияет на цену.
## 2. Постройте график размаха (boxplot) для анализа влияния состояния дома на его стоимость. График должен продемонстрировать, как состояние дома связано с его ценой и выявить возможные отклонения.
## 3. Постройте график размаха (boxplot), чтобы исследовать, как количество этажей влияет на стоимость недвижимости. График должен показать, как ценовые данные варьируются в зависимости от этажности.
## 4. Постройте график размаха (boxplot), чтобы проанализировать, как наличие вида на набережную влияет на стоимость недвижимости. График должен показать, есть ли значительное влияние наличия вида на цену.
## 5. Постройте диаграмму рассеяния, чтобы исследовать зависимость между площадью жилой зоны и стоимостью недвижимости. График должен показать, как площадь жилой зоны коррелирует с ценой.


```python
plt.figure(figsize=(8, 6))
plt.scatter(df['bedrooms'], df['price'], alpha=0.5)
plt.title('Цена недвижимости в зависимости от количества спален')
plt.xlabel('Количество спален')
plt.ylabel('Цена у.е.')
plt.grid(True)
```

## Вывод: нет линейной зависимости между количеством спален и стоимостью дома. Видно вылетающее значение >30.

```python
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['condition'], y=df['price'], whis=1.5)
plt.title('Цена недвижимости в зависимости от состояния')
plt.xlabel('Состояние дома')
plt.ylabel('Цена у.е.')
plt.grid(True)
plt.show()
```

## Вывод: с улучшением состояния дома растет цена. Наблюдаются вылеты многочисленные по цене, начиная от нормального состояния дома.

```python
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['floors'], y=df['price'], whis=1.5)
plt.title('Цена недвижимости в зависимости от количества этажей')
plt.xlabel('Количество этажей')
plt.ylabel('Цена у.е.')
plt.grid(True)
plt.show()
```

## Вывод: цена возрастает с увеличем количества этажей до 2.5. Большой разброс по цене для 2-этажных домов, однако медианная стоимость - около 500000 у.е.

```python
df.boxplot(column='price', by='waterfront', grid=False)
plt.title('Цена недвижимости в зависимости от наличия вида на набережную')
plt.suptitle('') # Удаление дополнительного заголовка, добавляемого pandas
plt.xlabel('Вид на набережную')
plt.ylabel('Цена у.е.')
plt.xticks([1, 2], ['Нет', 'Да'])
plt.grid(True)
```

## Вывод: большой разброс по цене у домов без вида на набережную, но медианная стоимость не более 500000 у.е. С наличием вида на набережную значительно растет цена дома.

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['sqft_living'], df['price'], alpha=0.5)
plt.title('Цена недвижимости в зависимости от площади жилой зоны')
plt.xlabel('Жилая площадь (sqft)')
plt.ylabel('Цена у.е.')
plt.grid(True)
```

## Вывод: существует прямая зависимочть между жилой площадью дома и его ценой. Чем больше площадь, тем выше цена.
